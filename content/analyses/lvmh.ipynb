# Executive Summary

LVMH Moët Hennessy Louis Vuitton SE (“LVMH”) is the world's largest luxury goods conglomerate, with over 75 prestigious brands spanning fashion, leather goods, wines & spirits, cosmetics, jewelry, and retail. The global luxury industry is entering a challenging phase; after a post-pandemic boom, 2024 saw the first contraction in luxury goods sales since 2008. LVMH has not been immune: its revenue dipped ~4% in the first half of 2025 and its fashion/leather goods segment (Louis Vuitton, Dior, etc.) saw sales decline for multiple quarters. Consequently, LVMH's share price has pulled back sharply from 2021-2022 highs, underperforming ultra-high-end peers like Hermès. However, LVMH's fundamental strengths - a portfolio of iconic brands with pricing power, global scale, and robust finances - position it well to weather short-term headwinds and capitalize on long-term luxury demand growth. The luxury sector's core drivers (wealth growth, status-driven consumption, Chinese demand) remain intact despite a near-term slowdown. LVMH currently trades at valuation multiples below its historical averages and below certain peers, suggesting potential upside as conditions normalize. We recommend a long-term bullish stance: LVMH's recent weakness offers an attractive entry for investors, though near-term caution is advised. In the short run, heightened volatility around events like Paris Fashion Week (Sep 29 to Oct 7, 2025) and the Q3 2025 sales release is expected. Long-term investors can consider accumulating shares on dips, while monitoring key risks (e.g. sustained luxury demand slowdown in the U.S./China, currency and tariff impacts).

Overall, LVMH's unparalleled brand assets and strong financial position underpin a favorable risk/reward balance, making it a compelling play on the enduring growth of the global luxury market. Below we present a detailed analysis of the industry, company, competition, investment thesis, and technical outlook, followed by our strategic recommendation.

# Industry Analysis

## Market Size and Growth

The personal luxury goods market (fashion, jewelry, etc.) reached approximately \$462 billion in 2024 after a post-Covid surge, but it contracted by about 2-3\% in 2024 amid economic uncertainties. In fact, 2024 - 2025 marks the luxury sector's first real slowdown since 2008 (aside from the brief pandemic shock). Bain & Company projects that under the most likely scenario, the luxury market will shrink by 2-5% in 2025 as the industry works through this downturn. Key macro headwinds (inflation, weaker post-pandemic consumer confidence, and geopolitical tensions) have curbed high-end spending in major markets. For example, rising economic uncertainty has shifted some consumer spending toward essentials and away from discretionary luxury purchases. Despite this near-term dip, the longer-term outlook remains positive: over the next 5 years, over 300 million new consumers (especially Gen Z and Asian middle class) are expected to enter the luxury market, and the number of high-net-worth individuals globally is projected to rise ~20%, expanding the customer base for luxury brands. Industry forecasts see the market resuming growth beyond 2025, reaching ~$700 billion by 2033 (about 5-7% CAGR), underpinned by increasing wealth in emerging markets.

## Geographic Trends

Asia-Pacific has been the engine of luxury growth, now accounting for \~35% of global luxury sales. China in particular drove much of the 2010s growth. However, Asia's dominance paused in 2023-2024 as China's demand cooled due to macroeconomic slowdowns and lingering travel restrictions. Early 2025 saw luxury sales in China dip \~11% year-on-year. Japan's luxury spending, which spiked in 2024 thanks to a weak yen attracting tourists, fell back to normal levels in early 2025. The United States, the other key market (\~25-30% of luxury sales), is also softening: U.S. consumers are becoming more cautious amid recession fears, and a stronger dollar plus price hikes by European brands have curbed some demand. That said, tourism and emerging markets provide bright spots. Europe saw solid local demand and tourist spending in early 2025, and markets like the Middle East, Mexico and Southeast Asia are thriving with affluent local buyers and inbound tourists. Luxury companies are increasingly investing in these high-growth regions (ex. opening stores in Vietnam, UAE, India) to diversify geographic exposure. In summary, the U.S. and China are cooling in the short term, while other regions (ex. Middle East, Latin America) are picking up some slack. Long-term, China and Asia at large are still expected to drive a huge portion of new luxury consumers, so their importance persists despite current weakness.

## Consumer Preferences and Trends

The luxury consumer is evolving. Generation Z and young millennials are becoming a significant segment of luxury buyers, but they exhibit different behaviors than older cohorts. Notably, Gen Z consumers are more skeptical of traditional luxury “brand narratives” and frequent price increases. Their net promoter scores for luxury brands are significantly lower than that of millennials, indicating lower loyalty and higher expectations for authenticity and value. Young shoppers crave both self-expression and individuality and align with trends like sustainability and social impact. This has given rise to several trends:

### Experiential Luxury

There is a shift from owning goods to seeking experiences. Segments like luxury hospitality, dining, travel, and even high-end wellness continue to grow faster than personal luxury goods. Affluent consumers (especially older generations) are increasingly spending on experiences (“memories over merchandise”), which can indirectly impact spending on handbags or apparel.

### Pre-Owned & Circular Fashion

The resale market for luxury is booming as younger consumers embrace second-hand luxury for sustainability and accessibility. Pre-owned luxury sales gained traction in recent years. This expands the customer ecosystem but also pressures brands to justify high prices for new items when perfectly good used items are available for less. Many luxury houses (including LVMH's brands) have responded by launching their own resale initiatives or repair services.

### Digital & E-commerce

Luxury lagged other sectors in e-commerce but has caught up since 2020. Online channels now approach \~25% of luxury sales. Social media and digital influencers heavily sway brand perception. However, the “digital fatigue” is real; many brands report slowing growth in online engagement (social follower growth down \~90% and engagement rates down 40% in recent analyses). Brands are thus focusing on experiential retail (unique in-store experiences, pop-ups, and personalization) to re-energize consumers.

### Taste for Variety vs. Loyalty

Luxury historically relies on brand loyalty (the heritage and status of a brand create repeat buyers). That loyalty is being tested. Consumers, especially aspirational middle-class buyers, have shown some “logo fatigue.” For instance, sales of big-logo handbags have slowed as shoppers opt for more timeless styles or even jewelry as an alternative investment in style. Top brands remain highly desirable, but there is evidence of rotating preferences. For instance, younger consumers might buy Gucci one season and Dior the next, driven by social media trends. Still, at the very high end, loyalty to houses like Hermès or Chanel remains strong (a Hermes buyer is unlikely to switch to another brand easily). Overall, there is a strong taste for novelty in terms of products (collaborations, new collections) but only a few mega-brands enjoy enduring loyalty. LVMH's ability to keep its brands “fresh” through creative directors and collaborations is critical to retaining loyalty while catering to the desire for variety.

## Competitive Structure and Barriers to Entry

The luxury industry features many brands but few dominant players. A handful of conglomerates (LVMH, Kering, Richemont, etc.) control a large share of the market, alongside powerful independent brands (Hermes, Chanel). In 2024, LVMH alone captured 24.3% of the global luxury market, far ahead of any competitor. The top 5 groups likely account for well over half of global luxury sales. Barriers to entry in true luxury are extremely high: it takes decades to build brand heritage and desirability. Newcomers struggle to attain the prestige of legacy houses. Moreover, the industry benefits from brand monopoly-like economics - each top brand (Louis Vuitton, Hermes, Cartier, etc.) is a category leader with loyal followings, making it hard for new brands to lure those customers. Other barriers include: access to the finest raw materials (ex. rare leathers, top vineyards - often controlled by incumbents), artisan craftsmanship know-how, and the capital for global retail networks and marketing. Consolidation has further raised barriers - giants like LVMH can outbid others for acquisitions and invest heavily in innovation and marketing. Additionally, luxury customers often trust established names, so breaking into the club is difficult. That said, competition within the sector is intense: rival brands compete on creativity and exclusivity rather than price. If consumer tastes shift (ex. a certain style falls out of favor), brands must adapt quickly or lose share. In summary, the luxury sector is an oligopoly at the top end, shielded by formidable barriers to entry (brand, heritage, capital), with moderate internal rivalry driven by fashion cycles and customer fickleness.

# Company Overview

## Business Portfolio & Revenue Streams

Overall, LVMH's 2024 revenue was €84.7 billion (a record high), but growth has stalled in 2023-25 compared to the ~20% annual growth rates seen in the 2021 rebound. Net profit in 2024 was roughly €14 billion (operating margin ~24%). In H1 2025, revenue was €39.8B and net profit €5.7B, down 22% YoY. Despite the dip, LVMH generated €4B in free cash flow in H1 (a +29% YoY jump) indicating strong cash generation even in a softer sales environment. The company's balance sheet is solid (net debt to EBITDA is modest, and it has a strong credit rating). It returned significant cash to shareholders (€13 per share dividend for 2024). These financials highlight LVMH's resilience: even with sales down a few percent, it maintains healthy profitability and cash flow, thanks to cost control and brand pricing power.

LVMH is a diversified luxury empire with 75 “Maisons” (houses) organized into five core divisions.

### Fashion & Leather Goods (42% of H1 2025 revenue)

This is LVMH's largest and most profitable segment. Flagship brands include Louis Vuitton, the world's biggest luxury fashion brand, as well as Christian Dior, Fendi, Celine, Loro Piana, Loewe, Marc Jacobs, Givenchy, and others. These brands produce luxury handbags, apparel, shoes, and accessories. In 2024, Fashion & Leather Goods (F&LG) generated over €40B in revenue and ~€15B in operating profit, contributing roughly 75% of LVMH's total profit.

Louis Vuitton is the crown jewel; known for its monogram handbags and 170-year heritage, it alone accounts for a large portion of LVMH's profits. Dior is another growth engine, especially in Asia. This segment has high margins (operating margin ~40% historically) thanks to strong pricing power. Recent trend: F&LG sales have stalled in 2023-25 after a post-Covid surge. In H1 2025, F&LG revenues were down 7%, reflecting weaker demand in the US and Asia (particularly among more price-sensitive “aspirational” shoppers). Nevertheless, local demand (ex. European and U.S. domestic customers) remained resilient, and the segment maintained a very high margin.

LVMH has been reinvesting in creativity, appointing star designers like Pharrell Williams (Louis Vuitton Men, 2023) and Jonathan Anderson (Dior, 2025) to reignite excitement. This segment's performance is crucial to LVMH's fortunes.

### Wines & Spirits (7% of H1 2025 revenue)

LVMH owns renowned champagne houses (Moet & Chandon, Veuve Clicquot, Dom Perignon), cognac (Hennessy, the world's top cognac), and other spirits (whisky, rum). This segment contributes smaller revenue (€5.2B in H1 2025) but has high profitability and brand cachet.

In H1 2025, Wines & Spirits revenue fell 7%. The major issue was weak cognac demand in the US and China amid trade tensions and consumer pullback. Hennessy, which relies on US and Chinese exports, saw volumes drop. On the positive side, champagne trends improved in Q2 2025 after a weak start. LVMH's champagne maisons even implemented strategic initiatives (new vintages, marketing tied to events like Formula 1) to sustain desirability.

Overall, this segment is sensitive to global trade conditions and consumer celebratory spending. Despite current weakness, its long-term outlook is solid; as emerging middle classes develop a taste for premium spirits, LVMH's brands stand to benefit.

### Perfumes & Cosmetics (10% of H1 2025 revenue)

This includes Parfums Christian Dior, Guerlain, Givenchy Beauty, Benefit Cosmetics, etc. These brands sell prestige fragrances, skincare, and makeup. Perfumes & Cosmetics had ~€4.1B revenue in H1 2025 (flat year-on-year).

The segment has lower margins than fashion, due to heavy marketing spend and competition from beauty giants, but it's an important gateway for younger/new customers (many consumers first access luxury via a Dior perfume or lipstick). In 2025, this division focused on “highly selective distribution” and innovation. Notably, Dior's Sauvage fragrance remains the world's best-selling men's fragrance, and Dior's makeup (Addict, Forever lines) and skincare have performed well, offsetting softness in some markets. Guerlain leveraged new product launches (ex. Abeille Royale serum) to drive growth.

This segment's stable performance (+0% in H1) shows the resilience of beauty even in a luxury downturn. It also serves strategic purposes: beauty products carry the brand image far and wide, and can yield strong loyalty (fragrance fans, etc.). LVMH's challenge is to grow this segment profitably amid fierce competition from the likes of Estee Lauder and L'Oreal.

### Watches & Jewelry (13% of H1 2025 revenue)

LVMH's brands here include Bulgari, TAG Heuer, Hublot, Chaumet, and since 2021, the iconic American jeweler Tiffany & Co. (acquired for $15.8B). In H1 2025, Watches & Jewelry revenue was roughly €5.1B (flat). Jewelry has been a relative bright spot in luxury; high-end jewelry demand has held up as some consumers view it as an investment (indeed, jewelry sales have outperformed handbags recently).

LVMH saw continued success with Tiffany & Co.: Tiffany's classic lines (ex. Heart, Key collections) and its revamped flagship store in New York have driven growth. Bulgari also shined with creative high jewelry collections (ex. Serpenti and new Polychroma collection) and high-profile exhibitions in Asia. The watches side is more challenged; TAG Heuer and Hublot face a soft market for luxury watches (and competition from Rolex, Omega, etc.). Still, they invest in partnerships (ex. TAG Heuer's Formula 1 tie-up) to stay relevant.

Overall, this segment's profit dipped in H1 2025 due to store renovation costs (especially the major Tiffany renovations). But strategically, LVMH's push into jewelry is paying off: jewelry is one of the fastest-growing luxury categories and less cyclical than fashion. LVMH is now a top-3 player in jewelry (with Richemont and Tiffany/Cartier).

### Selective Retailing (22% of H1 2025 revenue)

This includes LVMH's retail networks like Sephora (the world's largest luxury beauty retailer) and DFS (duty-free shops), plus boutiques like Le Bon Marche. In H1 2025, this segment had €8.6B revenue (up 2%) and notably was the only segment to grow and increase profit (+12% YoY). Sephora has been a standout, achieving growth in both revenue and profit. As travel rebounded, Sephora expanded globally (including returning to the UK market) and benefited from strong demand for beauty products. DFS (duty free) is still recovering - it relies on Asian travel flows (Chinese tourism), which haven't fully normalized.

But trends are improving as travel resumes. This retail segment diversifies LVMH's revenue and provides direct-to-consumer channels. It is lower-margin than product segments (Sephora's retail margin is single-digit), but recent performance shows it can bolster results even when core luxury sales slow. Impressively, in the luxury slowdown of 2024, LVMH’s selective retail division actually had strong momentum - Bain noted “strong growth momentum for LVMH's selective retailing unit (Sephora, DFS)” even as demand for core fashion weakened.

## Customer Base & Demand Drivers

LVMH's customer base spans from ultra high net worth elites to aspirational middle-class consumers, depending on the brand and product. For example, Louis Vuitton and Dior have broad appeal including wealthy professionals and upper-middle-class shoppers who save for a handbag. In contrast, Hermès (not LVMH-owned) caters more exclusively to ultra-rich clients. For LVMH, Asia (especially China) is critical - estimates suggest Chinese consumers (at home and abroad) historically accounted for 30-35% of LVMH's sales. The pandemic temporarily disrupted Chinese spend (due to travel halts), but domestic China luxury spending surged as a result. Now with travel resuming, Chinese tourists in Europe and Japan are spending again, albeit with some caution. The U.S. consumer is also vital (North America ~27% of revenue). LVMH's H1 2025 results noted “solid local demand in Europe and U.S.” but a decline in Japan (due to prior tourist surge) and flat trends in Rest of Asia. This indicates that American and European customers have so far been relatively resilient, whereas Asian markets have been softer recently.

Demographically, LVMH's brands increasingly target younger shoppers: the company has collaborated with artists (ex. Louis Vuitton x Yayoi Kusama, Jeff Koons; Dior x Travis Scott) and appointed younger creative directors to stay culturally relevant. Additionally, LVMH has a growing emphasis on female customers with high purchasing power, especially in jewelry and fashion. One notable development: Tiffany's integration likely broadened LVMH's reach among female fine jewelry buyers (a segment where Cartier/Hermes had more share historically). In summary, LVMH's revenue is broadly supported by affluent consumers worldwide, with Chinese, American, and European luxury shoppers being the main pillars, and the millennial/Gen Z cohort becoming an increasingly important driver (expected to contribute 70% of luxury purchases by 2025 according to industry studies).

## Competitive Advantages via The “7 Powers”

LVMH's competitive moat is extremely strong. Its strongest “powers” per Hamilton Helmer's framework are Brand (iconic heritage brands commanding customer love and pricing power) and Scale/Cornered Resources (global scale, control of unique assets). These create a self-reinforcing cycle: heritage brands yield high profits, which LVMH reinvests to further strengthen those brands and acquire new ones. As an illustration of brand power: in a 2024 analysis, Bain noted that 60% of luxury buyers under 35 prioritize heritage and craftsmanship - exactly the attributes LVMH's maisons embody - over trendy or generic products. This underpins why LVMH can endure short-term fashion swings; its brand equity is timeless.

### Branding

This is LVMH's greatest power. Its brands (Louis Vuitton, Dior, Fendi, Hennessy, etc.) carry immense emotional equity and prestige that allow premium pricing. In luxury, brand heritage and status are everything - LVMH's brands each have unique DNA and desirability cultivated over decades (or centuries). This creates a moat: consumers aspiring to a Louis Vuitton bag won't be easily swayed by a new label. The brand power also insulates LVMH from commoditization; for example, Louis Vuitton can raise prices regularly and still see demand (though recent reports suggest overuse of price hikes can alienate some shoppers). The intangible asset of brand is something only a few companies in the world (like LVMH, Hermes, Chanel) truly wield at this scale. Helmer's framework would squarely categorize this as cornered resource (heritage brands that can't be easily replicated) and branding power itself.

### Scale Economies

LVMH benefits from scale in operations and marketing. With €85B in sales, it can afford the best store locations, invest in cutting-edge marketing (ex. sponsoring major art exhibits, exclusive events), and negotiate favorable terms with suppliers. Its large retail network (6,300 stores worldwide) provides an extensive direct-to-consumer reach that smaller rivals can't match. Scale also helps spread risk - a slump in one brand or region can be offset by others. In advertising, LVMH's conglomerate structure means it can purchase media more efficiently or cross-promote brands in its ecosystem (ex. Champagne served at Louis Vuitton events). Process power is also evident in how LVMH efficiently acquires and integrates brands, leveraging a common pool of managerial talent and financial discipline that smaller fashion houses often lack. These scale advantages make LVMH a more formidable competitor, though in luxury the benefit is somewhat balanced by the need to keep brands distinct (it's not a pure cost-synergy play as in mass retail).

### Counter-Positioning

LVMH was a pioneer in the conglomerate model in luxury. In the 1980s-90s, while others focused on single brands, LVMH acquired numerous brands to build a multi-portfolio group. This counter-positioned it relative to traditional family-run luxury houses by providing diversified revenue streams and investment capital to grow brands globally. Competitors like Kering and Richemont later emulated this model. Today, LVMH's scale and diversity itself is a competitive advantage that newcomers can hardly replicate - a new entrant cannot easily counter-position against LVMH's breadth of offerings (from champagne to jewelry).

### Switching Costs

In luxury, switching costs for consumers are relatively low - buyers can easily choose a Gucci bag over a Louis Vuitton bag next time. However, one could argue there are psychological switching costs: loyalty programs (ex. Louis Vuitton VIC programs for valued clients), emotional attachment to a brand's heritage, and a collection effect (someone who has several Dior items may prefer to continue with Dior). But overall, LVMH doesn't rely on technical switching costs like a software company would. Instead, it uses customer loyalty and brand communities to retain clients. The high desirability of its products and the social cachet of owning them act as a soft barrier - if you deeply identify with being an LV customer, you are less likely to “switch” to a lesser-known brand.

### Network Effects

There are limited traditional network effects in luxury (one person owning a Louis Vuitton doesn't inherently make it more valuable for another to own one, unlike social networks). However, there is a social network effect of sorts: the more people aspire to LVMH brands, the more those brands are present in pop culture and social media, reinforcing their desirability. For example, seeing influencers and peers carrying an LV bag can increase its allure. LVMH also has an internal network effect in its ecosystem; ex. Sephora's strength as a retailer attracts more beauty brands to distribute through it, which in turn brings more customers, and so on. But by and large, network effects are not the primary power here beyond brand ubiquity reinforcing itself.

### Cornered Resource

LVMH has several. It owns exclusive vineyards (ex. Château d'Yquem, top Champagne vineyards) - resources that are physically limited and prestigious. It also “owns” creative talent - for instance, LVMH's ability to attract top designers (ex. Virgil Abloh, Kim Jones, Nicolas Ghesquiere) and maintain long-term relationships with artisan workshops (leather craftsmen, watchmakers) is a cornered resource. Competitors can't easily poach a centuries-old wine estate or a storied Parisian atelier under LVMH's umbrella. Additionally, its massive art collection and historical archives for brands like Dior are unique assets that new entrants lack. These cornered resources contribute to the authenticity and exclusivity of LVMH products, something very hard for a new brand to acquire quickly.

### Process Power

Over decades, LVMH has honed processes for luxury brand management - balancing heritage and innovation, scaling distribution without diluting exclusivity, etc. For example, LVMH's process of rotating creative directors while maintaining brand identity is an internal capability. The company is also known for operational excellence in areas like supply chain (ex. LVMH's logistics to ensure store restocks without flooding market) and its ability to execute large-scale events (like fashion shows, exhibitions) flawlessly. This organizational know-how acts as a competitive advantage that is not readily visible to consumers but results in superior performance and consistency.


# Competitive Landscape

LVMH leads a luxury sector that is oligopolistic, with a few giants and some highly-focused independent firms. Its competitive advantages (brand portfolio, financial might) are unmatched, yet the company is not invincible - recent trends show pockets of weakness that competitors are exploiting, such as Hermès capturing handbag share and high jewelry growth. Still, LVMH's diversification often means that when one area slows, another picks up (e.g. Sephora, Tiffany). Going forward, watch for whether LVMH's refreshed fashion offerings and marketing will outpace Kering's turnaround and sustain its dominance over Hermès/Richemont in the overall luxury arena.

## Key Competitors

The primary publicly-traded rivals are Kering (owner of Gucci, Saint Laurent, Bottega Veneta, Balenciaga, etc.), Richemont (owner of Cartier, Van Cleef & Arpels, Piaget, and Yoox/Net-a-Porter), and Hermes International (focused on the Hermes brand). Chanel is another major player (privately held by the Wertheimer family), with ~$10B+ annual revenue; while not conglomerate, it's a single-brand powerhouse. Other notable competitors include Prada Group (Prada, Miu Miu), Burberry, Swatch Group (watches like Omega, Blancpain) and Estee Lauder/L'Oreal in the beauty domain. However, none match LVMH's breadth and scale.

## Market Share

LVMH is the clear market leader with roughly 1/4 of the global luxury market by revenue in 2024. Kering and Richemont each have about 5-6% share of the luxury market (with €17B and €21B annual sales respectively in 2024). Hermes, with €15B in 2024 sales (growing strongly), has ~3-4% share.

In essence, LVMH's sales are greater than those of Kering, Richemont, and Hermès combined, underscoring its dominance. The conglomerate model has allowed LVMH to capture outsized share: by owning many top brands, it “crowds out” competitors at the high end. For example, in the crucial fashion leather goods segment, LVMH's Louis Vuitton and Dior together likely command a higher share than any rival's brands (Gucci, the main Kering brand, is one top competitor but its sales are about one-third of Louis Vuitton's). That said, certain niches have different leaders - ex. in hard luxury (watches/jewelry), Richemont with Cartier and Hermes (for jewelry) are formidable, and in beauty, L'Oreal (licenses for brands like Lancôme, etc.) is big. But overall, LVMH is number one or two in most categories it plays in.

## Competitive Dynamics

Recently, Hermes and Richemont have outperformed, whereas LVMH and Kering faced more challenges. Year-to-date 2025, Hermes's stock was roughly flat and Richemont slightly up, while LVMH's stock fell \~27% and Kering \~15%. This divergence stems from consumer mix and brand momentum. Hermes (and Cartier/Van Cleef under Richemont) cater to ultra-wealthy clients with timeless products (Birkin bags, high jewelry) - these customers were less affected by inflation or price fatigue, and actually, demand stayed strong (Hermes grew sales +15% in 2024 despite the market slowdown). In contrast, LVMH and Kering rely more on aspirational customers, who have been hit by high prices and economic worries. For instance, Gucci (Kering's flagship) has struggled badly, with sales down \~23% in early 2025 as its trendy streetwear wave faded and a brand revamp is in progress. LVMH's Louis Vuitton saw a more modest decline (\~6% down in Q2 2025), but it's still noteworthy given Louis Vuitton hadn't seen sales drops in years. This indicates that the competitive battle for the “aspirational” luxury consumer has intensified. Mid-tier luxury consumers have become more choosy, perhaps opting for classic investment pieces (jewelry, watches) over yet another pricey handbag. Consequently, Hermes and Cartier gained share, while Vuitton/Gucci saw softness.

## Strategy

### Product Strategy

LVMH's brands have introduced more entry-level offerings to capture younger clients - ex. Louis Vuitton launched a sub-$1,000 sneaker-ballerina shoe and expanded beauty products. This “democratization” carries risk (dilution of exclusivity), but aims to maintain volume. Hermes by contrast refuses to lower its price points or increase volume drastically, preserving exclusivity (and its legendary waitlists). Kering's Gucci is undergoing a design overhaul under a new creative director to recapture its trend leadership. Many brands are bringing in new creative talent - Dior at LVMH (hired J. Anderson), Gucci at Kering (Sabato De Sarno), Burberry (Daniel Lee) - betting that fresh designs will re-ignite demand. There is typically a lag of a few seasons before new designers' products hit stores, so H2 2025-2026 will be telling.

### Category Focus

LVMH has doubled down on jewelry (Tiffany) and beauty retail (Sephora) to diversify. Richemont remains focused on jewelry and watches (it divested most of its fashion brands). Kering, seeing weakness in its core, has moved into beauty too (it acquired fragrance brand Creed and secured the beauty license for Valentino in 2023) and recently invested in high-end fashion houses (ex. a 30% stake in Valentino) to broaden its portfolio beyond Gucci. Hermès sticks to what it does best - uber-luxury leather goods, silk and accessories, with limited diversification. These differing moves show LVMH's breadth as a competitive advantage - it can acquire and integrate across categories more readily than others (ex. Tiffany acquisition was a bold move few others could attempt).

### Financial Comparison

LVMH enjoys high margins but Hermes is the profitability champion of the industry (Hermes's operating margin exceeds 40%, vs ~26% for LVMH in 2023). This reflects Hermes's tight control on supply and perennial demand. Kering's margins have been shrinking (Gucci's issues), now closer to 20%. In valuation, Hermes commands a P/E ~47 (a hefty premium for its growth and resilience), while LVMH is around P/E ~21, below its peers like Hermès and even below Kering's ~23. Investors thus place a higher premium on Hermès' steady growth, whereas LVMH is seen as more cyclically exposed but also value-rich (multiple brands). Richemont's P/E is ~22-25. The lower multiple for LVMH (relative to Hermès) suggests the market expects slower growth or higher risk in LVMH's mix, but it also implies potential upside if LVMH proves its resilience.

### Market Share Shifts

As of 2024, LVMH remains the leader, but the gap narrowed slightly in segments like high-end leather goods where Hermès gained. Bain's data show LVMH's share actually ticked down from ~25% to 24.3% in 2024 due to the fashion slowdown. Meanwhile, Hermes and Richemont increased sales, implying share gains. If LVMH's major brands bounce back in 2025-26 (with new products, Chinese recovery, etc.), it could reclaim lost ground. The competitive landscape is thus in flux in the short term, with Hermès and Richemont currently on the offensive and LVMH and Kering working through a transitional period. All players face the broader challenge of re-energizing younger consumers and justifying their high prices. For example, price increases have been a double-edged sword - LVMH and others raised prices ~20-30% over 2021-2023, which boosted revenues but now some “aspirational” shoppers feel priced out, contributing to volume declines. Bain warned that brands must “rethink price hikes to win back disillusioned customers”. LVMH's scale allows it to be patient (it can afford a few soft quarters), whereas a smaller player might be more vulnerable in a downturn.

# Key Investment Thesis

## Drivers

Despite short-term headwinds, there are several compelling drivers for LVMH’s growth and stock appreciation.

### Unparalleled Brand Portfolio & Pricing Power

LVMH owns the top names in nearly every luxury vertical. Louis Vuitton, Dior, Tiffany, Moet, and others are brands with enduring appeal that can command premium prices and maintain customer loyalty even during downturns. This “fortress of heritage brands” has immense value. Each of these brands has strong pricing power - for instance, Louis Vuitton has historically raised prices ~1-2 times a year with minimal volume loss (until very recently) because of its status appeal. LVMH's brands benefit from timeless desirability: as CEO Bernard Arnault noted, they remain focused on “quality and desirability” and artisanal heritage to sustain customer demand. This brand equity underpins margins and distinguishes LVMH from ordinary retail companies. When luxury demand re-accelerates, LVMH's marquee brands are likely to be first in line to capture that upswing. In effect, owning LVMH stock gives exposure to a diversified basket of the world's most prestigious luxury assets - a unique value proposition that no other investment replicates.

## Data Collection and Preprocessing

```python
import yfinance as yf
import pandas as pd

# Fetch historical daily data for LVMH (MC.PA) in EUR
data = yf.download("MC.PA", start="2015-01-01", end="2025-08-01")
data = data[['Close']]  # focus on closing prices
print(data.head())      # verify data format
print(data.tail())
```

For modeling, we split the data into training and testing subsets:
* Training set: 2015-2024 data (to train the models).
* Testing set: Jan-Jul 2025 data (to evaluate model performance on recent unseen data).

We normalized the price series for the LSTM model using a Min-Max scaler (scaling to [0,1]) to help the neural network training converge. We then created sequences of the last 60 trading days to predict the next day's price. The choice of a 60-day window (about 3 months) balances including enough history for the LSTM’s memory without oversmoothing shorter patterns.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Split into train/test sets
train_data = data[data.index < "2025-01-01"]['Close']
test_data  = data[data.index >= "2025-01-01"]['Close']

# Scale prices to [0,1] range using training data stats
scaler = MinMaxScaler(feature_range=(0, 1))
train_vals = train_data.values.reshape(-1, 1)
test_vals = test_data.values.reshape(-1, 1)
scaler.fit(train_vals)
train_scaled = scaler.transform(train_vals)
test_scaled = scaler.transform(test_vals)

scaled_full = scaler.transform(data)

# Create sequences of 60 days for LSTM
seq_len = 60
X_train, y_train = [], []
for i in range(seq_len, len(train_scaled)):
    X_train.append(train_scaled[i-seq_len:i, 0])
    y_train.append(train_scaled[i, 0])
X_train = np.array(X_train).reshape(-1, seq_len, 1)
y_train = np.array(y_train)
print("Training sequences shape:", X_train.shape, "Target shape:", y_train.shape)
```

X_train is of shape (N, 60, 1) and y_train is the next-day normalized price. The test set will be used for validation after training.

## LSTM Modeling for Price Forecast

The network has 2 LSTM layers (50 hidden units each) and a fully connected output layer that produces a single price prediction. We use a mean-squared-error loss for regression and the Adam optimizer for training.

Traditional time-series models (like ARIMA) often struggle with non-linear patterns and long memory. LSTMs, however, can analyze complex patterns in historical data and retain long-term information via gated memory cells. This makes them ideal for financial time-series, capturing trends, cycles, and even some event-driven effects that span many days.

```python
import torch
import torch.nn as nn

# Define LSTM model architecture
class LSTMForecast(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=2):
        super(LSTMForecast, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc   = nn.Linear(hidden_size, 1)
    def forward(self, x):
        # LSTM returns outputs for all timesteps; we use the last output
        out, _ = self.lstm(x)                   # out: [batch, seq_len, hidden_size]
        last_out = out[:, -1, :]                # last time step output
        return self.fc(last_out)               # linear layer to get 1 value

# Initialize model, loss, optimizer
model = LSTMForecast(input_size=1, hidden_size=50, num_layers=2)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Train the LSTM model
epochs = 20
model.train()
for epoch in range(1, epochs+1):
    optimizer.zero_grad()
    outputs = model(torch.from_numpy(X_train).float())
    loss = criterion(outputs.view(-1), torch.from_numpy(y_train).float())
    loss.backward()
    optimizer.step()
    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Training Loss: {loss.item():.6f}")
```

We trained for 20 epochs; the loss steadily decreased, indicating the model was learning the data structure. To validate, we performed a one-step-ahead prediction on the test set (2025 data) and compared it to actual prices. In doing so, we fed the model the prior 60 days of actual data for each day in 2025, to simulate how it would perform in real time.

```python
model.eval()
start_idx = len(train_vals)
scaled_vals = scaled_full.flatten()
dates = data.index
preds, actuals, pred_dates = [], [], []
for i in range(start_idx, len(scaled_vals)):
    x_input = scaled_vals[i-seq_len:i].reshape(1, seq_len, 1)
    y_actual = scaled_vals[i]
    with torch.no_grad():
        y_pred = model(torch.from_numpy(x_input).float()).item()
    preds.append(y_pred)
    actuals.append(y_actual)
    pred_dates.append(dates[i])

pred_prices = scaler.inverse_transform(np.array(preds).reshape(-1, 1)).flatten()
actual_prices = scaler.inverse_transform(np.array(actuals).reshape(-1, 1)).flatten()

plt.figure(figsize=(12, 6))
plt.plot(pred_dates, actual_prices, label='Actual Price', linewidth=2)
plt.plot(pred_dates, pred_prices, label='LSTM Forecast', linestyle='--', color='orange')
plt.title("LVMH One-Step-Ahead Forecast vs Actual (2025)")
plt.xlabel("Date")
plt.ylabel("Price (EUR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The LSTM model demonstrated the ability to capture the general directional movement of LVMH's stock throughout 2025, especially its broad downward trajectory. As shown in the chart above, the forecasted prices (orange) track the actual trend (blue) reasonably well, though the model tends to underestimate the magnitude of sharp moves, particularly during volatile drops such as in March and April 2025. This behavior is consistent with LSTM models optimized for mean-squared error, which often produce conservative estimates during regime shifts or news-driven shocks. Despite this, the forecast maintains a steady signal with decent trend-following accuracy, and RMSE over the period remained relatively low compared to the €500-700 price range.

With this validation, we retrained the model on the full dataset up to August 1, 2025 and generated a forward-looking forecast from early August through mid-October. This forecast was produced via recursive prediction, where the model uses the last 60 days of actual prices to predict the next day, then uses that prediction in the next input window, continuing the process to simulate how the model might perform in a real-time setting leading up to major events like Q3 earnings and Paris Fashion Week.

```python
# Use the trained model to forecast future days
model.eval()
last_60_days = list(train_scaled.flatten()) + list(test_scaled.flatten())  # full data scaled
last_60_days = last_60_days[-60:]  # last 60 days up to Aug 1, 2025
last_60 = torch.tensor(last_60_days, dtype=torch.float32).view(1, -1, 1)

future_days = 50  # predict ~50 trading days ahead (Aug through early Oct 2025)
future_preds_scaled = []
with torch.no_grad():
    for _ in range(future_days):
        pred = model(last_60)
        future_preds_scaled.append(pred.item())
        # slide the window: drop oldest, add newest pred
        last_60 = torch.cat([last_60[:, 1:, :], pred.view(1,1,1)], dim=1)

# Inverse transform the scaled predictions back to price
future_preds_scaled = np.array(future_preds_scaled).reshape(-1, 1)
future_preds = scaler.inverse_transform(future_preds_scaled).flatten()

display(future_preds)
```

Future_preds is an array of predicted prices for the next 50 trading days (approximately up to mid-October 2025). We also prepared a corresponding date index for these predictions (skipping weekends), so we can plot them against actual dates.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import timedelta

# Assume `future_preds` contains predicted prices for next 50 trading days
# and `last_date` is the last available trading date from your dataset
last_date = pd.to_datetime("2025-08-01")
future_dates = pd.bdate_range(start=last_date + timedelta(days=1), periods=50)

# Create forecast series
forecast_series = pd.Series(future_preds.flatten(), index=future_dates)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(forecast_series, label="LSTM Forecasted Price", color='blue')

# Key events
plt.axvline(pd.to_datetime("2025-09-29"), linestyle='--', color='green', label='Paris Fashion Week')
plt.axvline(pd.to_datetime("2025-10-03"), linestyle='--', color='purple', label='Q3 2025 Earnings')

plt.title("LVMH Stock Price Forecast (LSTM Model)")
plt.xlabel("Date")
plt.ylabel("Predicted Price (EUR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The LSTM model's forward forecast from August 2025 through early October paints a stabilizing picture for LVMH stock. Rather than projecting continued bearish momentum, the model anticipates a gradual recovery, with prices forecasted to climb from ~€551 in early August to a plateau around €566 by mid-September. This suggests the model is detecting some exhaustion in the prior downtrend and expects a modest rebound, though still far below early 2025 highs.

Importantly, the model shows no sharp price spikes around either Paris Fashion Week (September 29 - October 7) or Q3 2025 earnings (October 3), indicating that absent unexpected external catalysts, the stock is likely to remain in a tight consolidation range. The predicted flatness into these events reflects the LSTM's reliance on past price behavior — if historically LVMH hasn't experienced large swings during these events, the model won't anticipate them.

This forecast highlights the limits of pattern-based prediction models: while the LSTM is effective at capturing momentum and smoothing out noise, it struggles to anticipate breakout volatility driven by qualitative factors like earnings surprises, product launches, or macro shifts. Therefore, while the model implies a steady-to-slightly-bullish baseline through the event window, traders should remain alert to narrative catalysts that could invalidate the technical trend.

## GARCH Volatility Forecasting

To complement the price forecast, we analyzed volatility using a GARCH(1,1) model on LVMH's daily returns. GARCH models are well-suited for financial time series because they capture time-varying volatility - the tendency of quiet and volatile periods to cluster. In other words, volatility comes in waves: calm markets stay calm until a shock hits, and turbulent markets tend to persist until risks subside. By modeling this, we can forecast if volatility is likely to increase around our event window (which could imply larger price swings).

First, we computed daily log returns (or percent returns) from the price data and inspected their statistical properties. LVMH's returns (like most stocks) show volatility clustering - stretches of low variance and bursts of high variance typically around news. We then fit a GARCH(1,1) model using the arch library in Python.

```python
!pip install arch
```

From the GARCH fit, we obtained the conditional volatility up to the last known date, as well as forecasts for the next 20 days. We annualized these for intuition (multiplying by √252).

```python
import numpy as np
from arch import arch_model

# Compute daily returns
returns = data['Close'].pct_change().dropna() *100 # percent returns
model = arch_model(returns, vol='Garch', p=1, q=1)
res = model.fit(disp="off")

# Get historical conditional vol (daily, in %)
garch_daily = res.conditional_volatility
garch_annualized = garch_daily * np.sqrt(252)  # annualize

# Forecast next 30 days
forecast_horizon = 30  # trading days (~1.5 months)
last_date = returns.index[-1]
forecast = res.forecast(horizon=forecast_horizon)

# Extract forecasted daily volatility (%)
forecast_vals = np.sqrt(forecast.variance.values[-1])  # daily std dev, %
forecast_annualized = forecast_vals * np.sqrt(252)     # annualize

forecast_index = pd.date_range(start=last_date + timedelta(days=1),
                               periods=forecast_horizon, freq='B')

forecast_vol = pd.Series(forecast_annualized, index=forecast_index)

# 20-day rolling std dev of returns, annualized
realized_vol = returns.rolling(window=20).std() * np.sqrt(252)

# Align both series to same start date
garch_annualized = garch_annualized[realized_vol.index]
realized_vol = realized_vol.dropna()
garch_annualized = garch_annualized.loc[realized_vol.index]

garch_full = pd.concat([garch_annualized, forecast_vol])

plt.figure(figsize=(14,6))
plt.plot(realized_vol, label='Realized 20-day Vol (%)', color='royalblue')
plt.plot(garch_full, label='GARCH Conditional Vol (%)', color='darkorange')

# Add vertical event lines
plt.axvline(pd.to_datetime("2025-09-29"), color='green', linestyle='--', label='Paris Fashion Week')
plt.axvline(pd.to_datetime("2025-10-03"), color='purple', linestyle=':', label='Q3 2025 Earnings')

# Add forecast dot
plt.scatter(pd.to_datetime("2025-10-03"), forecast_vol[-1], color='darkgreen', label='Forecast Vol (%)', zorder=5)

plt.title("LVMH Volatility: 2024–2025 (Realized vs GARCH Extended Forecast)")
plt.ylabel("Volatility (% annualized)")
plt.xlabel("Date")
plt.grid(True)
plt.xlim(pd.to_datetime("2024-01-01"), pd.to_datetime("2025-10-10"))
plt.legend()
plt.tight_layout()
plt.show()

# Final forecast output
print("1-day ahead vol forecast (%, annualized):", round(forecast_vol.iloc[0], 2))
```

The fitted GARCH(1,1) had typical parameters: a high persistence ($\alpha + \beta$ close to 1) indicating that volatility is mean-reverting but with long memory. In plain English, if volatility spikes due to an event, it will gradually decay back to normal, but that decay can take some time. Our GARCH model's implied half-life of volatility shocks was on the order of several days.

As of early August 2025, LVMH's annualized daily volatility was around ~30% (meaning a standard deviation of ~2% daily moves) - elevated compared to early 2025 when it was closer to 20%. This jump reflects the turbulent summer (post-H1 earnings).

The GARCH model predicts that volatility will moderate slightly going into September if no new shocks occur. The 1-day ahead volatility forecast is a bit lower than current realized vol, and over a 2-3 week horizon, it trends down toward a long-run average (~20% annualized). This is because, in absence of new disturbances, the model expects the recent volatility spike to dissipate.

However, crucially, the model does not incorporate the knowledge of upcoming events. A big earnings surprise on Oct 3 could cause a volatility surge beyond the model's baseline forecast. Historically, earnings releases have indeed caused volatility to jump for a few days. Our model can be interpreted as forecasting the baseline volatility if the events are in line with expectations. Should Q3 results or Fashion Week news shock the market, realized volatility will exceed the GARCH baseline (something an event-driven adjustment could address, but that's outside our current scope).

LVMH's historical volatility profile reveals a clear pattern of spikes around earnings announcements, followed by mean reversion. This behavior is evident in the realized volatility series — particularly the surges in late April 2025 (likely around Q1 earnings) and late July 2025 (post-H1 results). The GARCH model's conditional volatility series, while smoother, broadly tracks this cyclical pattern.

As shown in the figure above, the GARCH-forecasted volatility remains elevated leading into Q3 2025 earnings and Paris Fashion Week, hovering around 30% annualized. This aligns with the realized volatility spikes observed in previous event windows (ex. October 2024 and April 2025), underscoring the seasonal and event-driven nature of LVMH's stock dynamics. Despite strong fundamentals and technical resilience, the forecast suggests heightened short-term risk, potentially creating entry opportunities if prices dip into undervalued territory.
